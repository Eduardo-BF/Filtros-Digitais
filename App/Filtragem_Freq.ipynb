## Importando Sinal


```python
%%capture
!pip install numpy pandas matplotlib;

import pandas as pd
import matplotlib.pyplot as plt
```

Para sinais com Referência de tempo

```python
sinal_c = pd.read_csv("Dados/sinal_c.tempo.csv") 

# Plotar o sinal
plt.figure(figsize=(10, 3))
plt.plot(sinal_c["Time (s)"],sinal_c["Signal"], label="Sinal Original")
plt.xlabel("Tempo (s)")
plt.ylabel("Amplitude")
plt.title("Sinal composto por 2 Hz e 20 Hz")
plt.show()
```

Para sinais sem referência de tempo

```python
sinal_s = pd.read_csv("Dados/sinal_s.tempo.csv") 

plt.figure(figsize=(10, 3))
plt.plot(sinal_s["Signal"], label="Sinal Original")
plt.xlabel("Amostras")
plt.ylabel("Amplitude")
plt.title("Sinal composto por 2 Hz e 20 Hz")
plt.show()
```
