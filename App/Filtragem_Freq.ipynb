## Importando Sinal


```python
%%capture
!pip install numpy pandas matplotlib

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Para sinais com Referência de tempo

```python
sinal_c = pd.read_csv("Dados/sinal_c.csv") 

# Plotar o sinal
plt.figure(figsize=(10, 3))
plt.plot(sinal_c["Time (s)"],sinal_c["Signal"], label="Sinal Original")
plt.xlabel("Tempo (s)")
plt.ylabel("Amplitude")
plt.title("Sinal composto por 2 Hz e 20 Hz")
plt.show()
```

##### Para sinais sem referência de tempo

```python
sinal_s = pd.read_csv("Dados/sinal_s.tempo.csv") 

plt.figure(figsize=(10, 3))
plt.plot(sinal_s["Signal"], label="Sinal Original")
plt.xlabel("Amostras")
plt.ylabel("Amplitude")
plt.title("Sinal composto por 2 Hz e 20 Hz")
plt.show()
```

## Filtro

```python

# Convoluir o sinal com o filtro FIR manual
sinal = sinal_c["Signal"]
filt_coeffs_LOW  = [1,-1,1,-1,1]
filt_coeffs_HIGH = [1,-1,-1,1]

filt_sinal_LOW = np.convolve(sinal, filt_coeffs_LOW, mode='same')
filt_sinal_HIGH = np.convolve(sinal, filt_coeffs_HIGH, mode='same')

# Plotando o sinal original e o sinal filtrado
plt.figure(figsize=(10, 6))
plt.subplot(3, 1, 1)
plt.plot(sinal)
plt.title("Sinal Original")
plt.subplot(3, 1, 2)
plt.plot(filt_sinal_LOW)
plt.title("Sinal Filtrado (Passa Baixas)")
plt.subplot(3, 1, 3)
plt.plot(filt_sinal_HIGH)
plt.title("Sinal Filtrado (Passa Altas)")
plt.tight_layout()
plt.show()
```
